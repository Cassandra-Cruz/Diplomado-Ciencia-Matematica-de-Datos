# Dataset description

###### This dataset is having the data of 1K+ Amazon Product's Ratings and Reviews as per their details listed on the official website of Amazon (https://www.kaggle.com/datasets/karkavelrajaj/amazon-sales-dataset).

Features:

*   ***product_id*** - Product ID
*   ***product_name*** - Name of the Product
*   ***category*** - Category of the Product
*   ***discounted_price*** - Discounted Price of the Product
*   ***actual_price*** - Actual Price of the Product
*   ***discount_percentage*** - Percentage of Discount for the Product
*   ***rating*** - Rating of the Product
*   ***rating_count*** - Number of people who voted for the Amazon rating
*   ***about_product*** - Description about the Product
*   ***user_id*** - ID of the user who wrote review for the Product
*   ***user_name*** - Name of the user who wrote review for the Product
*   ***review_id*** - ID of the user review
*   ***review_title*** - Short review
*   ***review_content*** - Long review
*   ***img_link*** - Image Link of the Product
*   ***product_link*** - Official Website Link of the Product

##### First, we write the discounted prices as integers, then we remove the Indian rupee symbol and convert the values to float so that we can do the conversion to dollars (*0.12) and generate a new column with those price calculations.

```python
df['discounted_price'] = df['discounted_price'].apply(lambda x: x.replace(',',''))
```

```python
df['discounted_price'] = df['discounted_price'].apply(lambda x: x.replace('₹',''))
```

```python
df['discounted_price'] = df['discounted_price'].astype(float)
```

```python
df['discounted_price_dlls'] = df['discounted_price'] * 0.12
df.head()
```

##### We perform the above steps to convert 'actual_price' to dollars and create a new column.

```python
df['actual_price'] = df['actual_price'].apply(lambda x: x.replace(',',''))
df['actual_price'] = df['actual_price'].apply(lambda x: x.replace('₹',''))
df['actual_price'] = df['actual_price'].astype(float)
df['actual_price_dlls'] = df['actual_price'] * 0.12
df
```

```python
def format_title(title, subtitle=None, subtitle_font_size=14):
        title = f'{title}'
        if not subtitle:
            return title
        subtitle = f'{subtitle}'
        return f'{title}{subtitle}'
```
